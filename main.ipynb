```python
import matplotlib.pyplot as plt

# --- 1. EKRAN GÖRÜNTÜSÜNDEN ALINAN VERİLER ---
steps = [100, 200, 300, 400, 500, 600]
training_loss = [0.411900, 0.389700, 0.356600, 0.348500, 0.350700, 0.304400]
validation_loss = [0.382805, 0.369552, 0.366814, 0.362549, 0.361882, 0.367852]

# --- 2. GRAFİK AYARLARI (Profesyonel Görünüm) ---
plt.figure(figsize=(10, 6), dpi=120) # Yüksek çözünürlük

# Training Loss Çizgisi (Mavi)
plt.plot(steps, training_loss, label='Training Loss', color='#1f77b4', linewidth=2.5, marker='o', markersize=6)

# Validation Loss Çizgisi (Kırmızı)
plt.plot(steps, validation_loss, label='Validation Loss', color='#d62728', linewidth=2.5, marker='s', markersize=6)

# --- 3. DETAYLAR VE ETİKETLER ---
plt.title('Model Training (Training and Validation Loss)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('(Steps)', fontsize=12)
plt.ylabel('Losses', fontsize=12)
plt.legend(fontsize=11, loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)

# En iyi noktayı işaretleyelim (Step 500 - En düşük Val Loss)
best_step = 500
best_val = 0.361882
plt.annotate(f'Best point\n(Val: {best_val:.4f})', xy=(best_step, best_val), xytext=(best_step-50, best_val+0.02),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10, color='darkgreen')

# --- 4. KAYDETME ---
output_file = "loss_graph_deep.png"
plt.savefig(output_file, bbox_inches='tight')
print(f"✅ Grafik oluşturuldu ve kaydedildi: {output_file}")

# Ekrana bas
plt.show()
```
